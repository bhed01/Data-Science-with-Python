## DataFrame.plot

```python
np.random.seed(123)

df = pd.DataFrame({'A': np.random.randn(365).cumsum(0),
                   'B': np.random.randn(365).cumsum(0)+20,
                   'C': np.random.randn(365).cumsum(0)-20},
                  index = pd.date_range('1/1/2017', periods=365))
df.head()
```

```python
df.plot();
```

```python
df.plot('A', 'B', kind='scatter');
```

```python
df.plot.scatter('A', 'C', c='B', s=df['B'], colormap='viridis');
```

```python
ax = df.plot.scatter('A', 'C', c='B', s=df['B'], colormap='viridis')
ax.set_aspect('equal');
```

```python
df.plot.box();
```

```python
df.plot.hist(alpha=0.5);
```

```python
df.plot.kde();
```

# Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook
```

```python
np.random.seed(1234)

v1 = pd.Series(np.random.normal(0, 10, 1000), name='v1')
v2 = pd.Series(2*v1 + np.random.normal(60, 15, 1000), name='v2')
```

```python
plt.figure()
plt.hist(v1, alpha=0.5, bins=np.arange(-50, 150, 5), label='v1')
plt.hist(v2, alpha=0.5, bins=np.arange(-50, 150, 5), label='v2')
plt.legend();
```

```python
plt.figure()
plt.hist([v1, v2], histtype='barstacked', density=True);
v3 = np.concatenate((v1, v2))
sns.kdeplot(v3);
```

```python
sns.jointplot(x=v1, y=v2, alpha=0.4);
```

```python
grid = sns.jointplot(x=v1, y=v2, alpha=0.4);
grid.ax_joint.set_aspect('equal')
```

```python
sns.jointplot(x=v1, y=v2, kind='hex');
```

```python
sns.set_style('white')
```

```python
sns.jointplot(x=v1, y=v2, kind='kde', space=0);
```

```python
iris = pd.read_csv('./data/iris.csv')
sns.pairplot(iris, hue='species', diag_kind='kde')
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(121)
sns.swarmplot(x='species', y='petal_length', data=iris)
plt.subplot(122)
sns.violinplot(x='species', y='petal_length', data=iris)
```

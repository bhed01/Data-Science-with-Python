# Air Quality Index (India)

```python
import pandas as pd

%matplotlib notebook
```

```python
aqi = pd.read_csv('./data/city_day.csv')
aqi.head()
```

```python
#Extracting only delhi record
aqi = aqi[aqi['City']=='Delhi']
aqi = aqi[['Date', 'PM10', 'NO', 'NO2', 'NH3',  'PM2.5', 'O3', 'SO2', 'AQI', 'CO']]

#conveting date to pandas datetime index
aqi['Date'] = pd.to_datetime(aqi['Date'])
aqi.set_index('Date', inplace=True)
```

```python
#plotting the line graph of data
fig = aqi.resample('M').mean().plot(title='Air Quality Index of Delhi, India (2015-2020)',
                                      style=['--']*7+['-', '--'],
                                      figsize=(10, 6),
                                      colormap='nipy_spectral',
                                      ylim=[0, 500])
#setting grid
fig.grid(axis='both', alpha=0.5)

#removing border
for spine in fig.spines.values():
    spine.set_visible(False)

#setting x axis tick labels
fig.set_xticklabels('')
fig.set_xticks(['2015-07', '2016-07', '2017-07', '2018-07', '2019-07', '2020-07'], minor=True)
fig.set_xticklabels(['2015','2016','2017','2018','2019', '2020'], minor=True)
fig.tick_params(left=False, bottom=False)
fig.tick_params(which='minor', bottom=False)

fig.set_ylabel('Air Qullity Index');
```
